```python
from PIL import Image
import os
```

```python
# Define the paths to the source and destination directories
source_dir = "E:/JAIN/PCL/Plant_Anomaly_Detection/data_augmented/dataset"
destination_dir = "E:/JAIN/PCL/Plant_Anomaly_Detection/resizing_images"
new_size = (512, 512)  # Set a larger size for better clarity
```

```python
# Traverse the directory structure and resize images
for plant_type in os.listdir(source_dir):
    plant_path = os.path.join(source_dir, plant_type)
    if os.path.isdir(plant_path):
        for condition in os.listdir(plant_path):
            condition_path = os.path.join(plant_path, condition)
            if os.path.isdir(condition_path):
                # Create corresponding directory in the destination folder
                dest_condition_path = os.path.join(destination_dir, plant_type, condition)
                os.makedirs(dest_condition_path, exist_ok=True)
                
                # Process each image in the folder
                for image_name in os.listdir(condition_path):
                    source_image_path = os.path.join(condition_path, image_name)
                    dest_image_path = os.path.join(dest_condition_path, image_name)
                    
                    try:
                        # Open the image, resize it, and save it to the destination folder
                        with Image.open(source_image_path) as img:
                            img_resized = img.resize(new_size, Image.LANCZOS)
                            img_resized.save(dest_image_path)
                        print(f"Resized and saved: {dest_image_path}")
                    except Exception as e:
                        print(f"Could not process {source_image_path}: {e}")

print("Image resizing completed.")
```
